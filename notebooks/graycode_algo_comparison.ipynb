```python
from pathlib import Path
import utils.benchmark as bm
import utils.profile as pf
import utils.profile_line as pl

NOTEBOOK_DIR = Path.cwd()
PROJECT_ROOT = NOTEBOOK_DIR.parent
ANALYSIS_DIR = PROJECT_ROOT / ".analysis"
ANALYSIS_DIR.mkdir(exist_ok=True)
```

```python
import cses.introductory_problems.gray_code_reflect_prefix as rp
import cses.introductory_problems.gray_code_bit_trick as bt

with (ANALYSIS_DIR / "graycode_bench.txt").open("w") as f:
    print("Gray Code Benchmark\n", file=f)

    for n in range(1, 17):
        data = str(n).encode().split()

        rp_time = bm.benchmark(rp.solve, data)
        bt_time = bm.benchmark(bt.solve, data)

        print(
            f"{n:2d}: "
            f"RP={rp_time:.6f}s  "
            f"BT={bt_time:.6f}s",
            file=f
        )
```

```python
pf.profile(rp.solve, data, file="graycode_rp_profile.txt")
pf.profile(bt.solve, data, file="graycode_bt_profile.txt")
```

```python
pl.line_profile(rp, rp.solve, data, file="graycode_rp_profile_line.txt")
pl.line_profile(bt, bt.solve, data, file="graycode_bt_profile_line.txt")
```
