# Algorithm Analysis

Workflow

Correct → Complexity → Benchmark → Profile → Optimize → Re-benchmark → New Algorithm?


## Imports


```python
import importlib
from timeit import timeit
import cProfile
import pstats
from line_profiler import LineProfiler
```

## Helpers


```python
def benchmark(func, data, number=3):
    return timeit(lambda: func(iter(data)), number=number) / number


def profile(func, data, iterations=1000):

    def bench():
        for _ in range(iterations):
            func(iter(data))

    profiler = cProfile.Profile()
    profiler.runcall(bench)

    stats = pstats.Stats(profiler)
    stats.strip_dirs()
    stats.sort_stats("tottime")
    stats.print_stats(20)


def line_profile(module, func, data):

    importlib.reload(module)

    lp = LineProfiler()
    lp.add_function(func)
    lp.runcall(func, iter(data))
    lp.print_stats()
```

## Experiment

```python
# Example:
import cses.introductory_problems.gray_code_reflect_prefix as rp
import cses.introductory_problems.gray_code_bit_trick as bt

for n in range(1, 17):
    data = str(n).encode().split()

    print(
        f"{n:2d}: "
        f"RP={benchmark(rp.solve, data):.6f}s  "
        f"BT={benchmark(bt.solve, data):.6f}s"
    )
```
