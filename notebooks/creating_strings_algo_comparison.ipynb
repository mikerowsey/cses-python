```python
from pathlib import Path
import utils.benchmark as bm
import utils.profile as pf
import utils.profile_line as pl

NOTEBOOK_DIR = Path.cwd()
PROJECT_ROOT = NOTEBOOK_DIR.parent
ANALYSIS_DIR = PROJECT_ROOT / ".analysis"
ANALYSIS_DIR.mkdir(exist_ok=True)
```

```python
inputs = ["abc", "aybabtu", "aaaaaaaa", "abcdefgh", "abz", "a"]
```

```python
import cses.introductory_problems.creating_strings_itertools as csit
import cses.introductory_problems.creating_strings_recursive as csrs

with (ANALYSIS_DIR / "creating_strings_bench.txt").open("w") as f:
    print("Creating Strings Benchmark\n", file=f)

    for n, data in enumerate(inputs):

        csit_time = bm.benchmark(csit.solve, data.encode().split())
        csrs_time = bm.benchmark(csrs.solve, data.encode().split())

        print(
            f"{n:2d}: "
            f"csit={csit_time:.6f}s  "
            f"csrs={csrs_time:.6f}s",
            file=f
        )
```
